```python
# repartition etablissement
nom_type_select = 'Restauration'

fig = plt.figure(figsize=(10,6))
my_df = pd.DataFrame(df[df['type_etablissement_extraction']==nom_type_select]['departement'].value_counts()[:20])

sns.barplot(data=my_df,
            x='departement',
            y=my_df.index,
            palette='Blues_d')
 
#st.pyplot(fig)
plt.title("Nombre d'etablissements\n("+nom_type_select+")")
plt.show()
```

```python
# Repartition POI

df_e = df['type_etablissement_extraction']
idx = df_e.value_counts()[df_e.value_counts()<20000].index
df_donut = df_e.apply(lambda x: x if x not in idx else 'Autres')
```

```python
my_circle = plt.Circle( (0,0), 0.7, color='white')
df_donut.value_counts(normalize=True).plot(kind='pie',ylabel=None);
p = plt.gcf()
plt.ylabel('')
p.gca().add_artist(my_circle)
plt.title("Repartition du nombre d'etablissements");
```
