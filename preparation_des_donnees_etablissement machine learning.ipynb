```python
 list_stopwords = nltk.corpus.stopwords.words("english")
```

```python
def clean_sw(sentence):
  # Minuslue:
  sentence = sentence.lower()

  for punctuation in string.punctuation:
    sentence = sentence.replace(punctuation, '')

  #spacy découpe automatiquement en tokens avec cette syntaxe :
  sent_tokens = nlp(sentence)

  #lemmatizing:
  tokens_lemm = [token.lemma_.lower() for token in sent_tokens]

  # stopwords:
  tokens_clean = []
 
  for words in tokens_lemm:
    if words not in list_stopwords:
      tokens_clean.append(words)
      
  # Transformation de la liste en string
  tokens_clean_sw = ' '.join(tokens_clean)

  return tokens_clean_sw
```

```python
# seulement sur hebergement et restauration
df_mdl = df_mdl[(df_mdl['type_etablissement_extraction']=='Restauration') | (df_mdl['type_etablissement_extraction']=='Hébergement')]
```

```python
X = df_mdl['descriptif_clean']
y = df_mdl['type_etablissement_extraction']
X_train,X_test,y_train, y_test = train_test_split(X, y, random_state=0, train_size = 0.75)
print("The length of the initial dataset is :", len(X))
print("The length of the train dataset is   :", len(X_train))
print("The length of the test dataset is    :", len(X_test))
```

```python
X_train
```

```python
vectorizer = CountVectorizer()
vectorizer.fit(X_train, y_train)

X_train_CV = vectorizer.transform(X_train)
X_test_CV = vectorizer.transform(X_test)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf  = TfidfVectorizer()
tfidf.fit(X_train)

X_train_CV = tfidf .transform(X_train)
X_test_CV = tfidf .transform(X_test)
```

```python
modelLR = LogisticRegression()
modelLR.fit(X_train_CV,y_train)
```

```python
y_pred_train = modelLR.predict(X_train_CV)
y_pred_test = modelLR.predict(X_test_CV)
```
