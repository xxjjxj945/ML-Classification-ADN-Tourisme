```python
def nuage_mots(etablissement):
    # word cloud 
    mask = df0['type_etablissement'].str.contains(etablissement) 

    text = " ".join(review.lower() for review in df0.loc[mask,'descriptif'].dropna())
    print ("There are {} words in the combination of all review.".format(len(text)))

    # Create and generate a word cloud image:
    wordcloud = WordCloud(stopwords=fr_stopwords, 
                          max_font_size=50, max_words=100,
                          background_color="white").generate(text)

    # Display the generated image:
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()
```

```python
df0['type_etablissement'] = df0['type_etablissement'].apply(eval)
```

```python
dico = {
    'Hébergement' : ['schema:Accommodation', 'schema:LodgingBusiness', 'Accommodation', 'RentalAccommodation', 'SelfCateringAccommodation'],
    'Restauration' : ['schema:FoodEstablishment', 'schema:Restaurant', 'FoodEstablishment', 'Restaurant'],
    'Site sportif, récréatif et de loisirs' : ['SportsAndLeisurePlace', 'ActivityProvider', 'LeisureSportActivityProvider'],
    'Site culturel': ['CulturalSite','schema:Event', 'schema:MusicEvent', 'Concert', 'CulturalEvent', 'EntertainmentAndEvent'],
    'Commerce de détail': ['schema:LocalBusiness', 'Store'],
    'Service d''information touristique': ['schema:TouristInformationCenter', 'LocalTouristOffice', 'TouristInformationCenter'],
    'Visite': ['olo:OrderedList', 'Tour', 'WalkingTour'],
    'Pratique':['schema:Product', 'Practice', 'Product'],
    'Site naturel':['schema:Landform', 'NaturalHeritage'],
    'Service pratique':['ConvenientService', 'PicnicArea'],
    'Prestataire de service':['ServiceProvider'],
    'Transport': ['Parking', 'Transport'],
    'Site d''affaires':['BusinessPlace'],
    'Autre':['PlaceOfInterest', 'PointOfInterest']
       }
```

```python
dico
```

```python
def extraction(liste):
    for key in dico:
        for nom in liste:
            if nom in dico[key]:
                return key
    return liste
```

```python
df0['type_etablissement_extraction'] = df0['type_etablissement'].apply(extraction)
```

```python
import folium
from folium.plugins import MarkerCluster
```

```python
# centrage de la carte sur la premiere adresse
mask0 = df0['type_etablissement'].str.contains('Restaurant') 
mask1 = df0['departement']=='Paris'
df_mask = df0.loc[mask0&mask1]

point = [ df_mask['lat'].iloc[0],df_mask['lng'].iloc[0] ]

width, height = 800, 400

fig = folium.Figure(width=width, height=height)

m = folium.Map(location=point,zoom_start=8, control_scale=True, width=width, height=height)
```

```python
#for name in ['Restaurant','Accomodation']:

mask0 = df0['type_etablissement'].str.contains('Restaurant') 
mask1 = df0['departement']=='Paris'
df_mask = df0.loc[mask0&mask1]


marker_cluster = MarkerCluster(
    name='Restaurant',
).add_to(m)

# affichage de tous les emplacements
for i in range(len(df_mask)):
    nom = df_mask['nom'].iloc[i]
    point = [ df_mask['lat'].iloc[i],df_mask['lng'].iloc[i] ]

    folium.Marker(location=point,
                  popup=nom).add_to(marker_cluster)
    
mask0 = df0['type_etablissement'].str.contains('Accommodation') 
mask1 = df0['departement']=='Paris'
df_mask = df0.loc[mask0&mask1]


marker_cluster = MarkerCluster(
    name='Accommodation',
).add_to(m)

# affichage de tous les emplacements
for i in range(len(df_mask)):
    nom = df_mask['nom'].iloc[i]
    point = [ df_mask['lat'].iloc[i],df_mask['lng'].iloc[i] ]

    folium.Marker(location=point,
                  popup=nom).add_to(marker_cluster)

folium.LayerControl().add_to(m)

fig.add_child(m)
```
